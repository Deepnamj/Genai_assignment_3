```python
def apply_discount(price,discount_percentage=5): # to give default value
    if discount_percentage < 0 or discount_percentage > 60:
        raise ValueError("Discount % cannot be less than 0% or greater than 60%")
    disc = price * (discount_percentage / 100)
    final_p = price - disc
    
    return final_p
#def apply_dis(price):
 #   discount = price * (5 / 100)
  #  final_price = price - discount
   # return final_price
print(apply_discount(1000,10))
print(apply_discount(500))
```

```python
def factorial(n):
    if n < 0:
        print("Factorial is not defined for -ve nos.")
        #exit()
        return None
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)  # recursive case: n! = n * (n-1)!
        #fact = 1
        #for i in range(n,1,-1): #range(start,stop,step)
        #    fact *= i #5! = 5*4*3*2*1
        #return fact
print(factorial(5))
print(factorial(0))
print(factorial(-3))
    
```

```python
def apply_gst(price):
    gst = lambda x: x * (18 / 100) #lambda syntax: lambda arguments: expression
    final_price = price + gst(price)
    return final_price
print(apply_gst(100))

def apply_dis_gst(price,disc_per,gst_per): #always calc disc before gst
    disc = lambda x : x * (disc_per / 100)
    disc_price = price - disc(price)
    gst_amount = lambda x : x * (gst_per / 100) 
    final_price = disc_price + gst_amount(disc_price)
    return final_price
print(apply_dis_gst(100, 10, 18))
```

```python
def process_prices(prices):
    disc_p = list(map(lambda x: x - (x * (10 / 100)), prices)) 
    #print(disc_p)      # 10% discount on all prices
    filtered_p = list(filter(lambda x: x > 300, disc_p))  # keep only prices above 300
    return disc_p, filtered_p  # return both lists

all_discounted, above_threshold = process_prices([100, 500, 900, 50, 750])
print("All discounted prices:", all_discounted) 
print("Prices above 300:    ", above_threshold) 
```

```python

#p_list = []

def add_price(p_list,price):
    if price < 0:
        print("Price cannot be -ve")
        return
    p_list.append(price)

def get_avg_p(p_list):
    return sum(p_list) / len(p_list)

def get_max_p(p_list):
    return max(p_list)

p_list = []
while True:
    print("1 - Add price")
    print("2 - Show avg price")
    print("3 - Get max price")
    print("q - Quit")
    choice = input("Enter your choice: ")
    if choice == '1':
        try:
            price = float(input("Enter price: "))
            add_price(p_list, price)
            print("Price added ")
        except:
            print("Invalid input")
    elif choice == '2':
        if len(p_list) == 0:
            print("No prices added")
        else:
            avg = get_avg_p(p_list)
            print("Avg price:", avg)
    elif choice == '3':
        if len(p_list) == 0:
            print("No prices added")
        else:
            max = get_max_p(p_list)
            print("Max price:",max)
    elif choice == 'q':
        print("Exited!!!")
        break
    else:
        print("Invalid choice")

```
